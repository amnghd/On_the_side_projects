```python
import pandas as pd
import re
import scipy
```

```python
purchases = pd.read_csv(r"data/recom/purchase_history.csv")
items =  pd.read_csv(r"data/recom/item_to_id.csv")
```

```python
column_dict = {r[1][1]:r[1][0] for r in items.iterrows()}
```

```python
df_list = []

for row in purchases.iterrows():
    index = row[0]
    user_id = row[1][0]
    items = row[1][1]
    items = list(map(int, re.split('[^\d]', items)))
    for item in items:
        df_dict = {}
        column = column_dict[item]
        df_dict['index']=index
        df_dict[column] = 1
        df_dict['user_id'] = user_id
        df_list.append(df_dict)

new_df = pd.DataFrame(df_list)
new_df.fillna(0,inplace=True)
```

Now we have all the items that each user has bought and for each item we have a list of users that have bough it.

### Problem 1:
Who is the customer that has bought the most number of items in her lifetime?

```python
# for this one we will contunie working on the purchase data frame
purchases['num_items'] = purchases.id.apply(lambda x: len(list(map(int, re.split('[^\d]', x)))))
```

```python
best_user = purchases.groupby('user_id')['num_items'].sum().sort_values(ascending=False).head(1)
```

```python
best_user_id = best_user.index.values[0]
best_user_count = best_user.values[0]
print('User with user id : {} bought highest number of products: {} number'.format(best_user_id, best_user_count))
```

### Problem 2:
Now we go toward higher granularity. We would like to know for each product, which customer bought the most.

```python
total_sales = new_df.drop('index', axis=1).groupby('user_id').sum()
```

```python
total_sales.head()
```

```python
print("User with the highest product most:")
print("Product\t\tUser id")
total_sales.idxmax().sort_values(ascending=False)
```

```python
## off to the clustering part
produts = total_sales.T
product_names = produts.index.values
```

```python
produts.head()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
sparse_df = csr_matrix(produts)
```

```python
# constants, after a few trials and error this one is chosen
n_dims = 48
```

```python
# reducing data dimentionality and normalizing
# we do normalizing becasue we dont want buyers who buy a lot 
# dominate our data
svd = TruncatedSVD(n_dims) # to work with sparse matrixes
normalizer = Normalizer(copy=False)
lsa = make_pipeline(normalizer, svd) # first normalize then PCA
reduced_products = lsa.fit_transform(sparse_df)
```

```python
explained_variance = svd.explained_variance_ratio_.sum()
# The svd explains only a tiny portion of the variance
print("Explained variance %fs" % (explained_variance))
```

```python
inertias = [] 
for i in range(1,15): # you can theoritically have more cluster (same as the number of products)
    kmeans = KMeans(n_clusters=i, random_state=23)
    kmeans.fit(reduced_products)
    inertias.append(kmeans.inertia_)
```

```python
# visualizing the inertias
plt.plot(range(1,15), inertias, marker='D')
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Inertia vs number of clusters")
```

12 clusters seem to be the sweet spot of this clustering algorithm.

```python
kmeans = KMeans(n_clusters=12, n_jobs=4, n_init = 50, precompute_distances=True, random_state=23)
clusters = kmeans.fit_predict(reduced_products)
```

```python
clustered_products = pd.DataFrame({'cluster':clusters, 'product':product_names})
clustered_products.sort_values('cluster')
```
