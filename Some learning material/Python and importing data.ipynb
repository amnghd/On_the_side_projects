# Python and Importing file locally

#### importing basic text files
- use context manager to open up a connection to the text file
- set mode='r' as you are reading the file
- make sure correct encoding is available
- there are three functions you can use to read the files in the context manager
    - read() which reads the whole text as a string.
    - readlines() which reads the whole text into array of lines
    - readline() which return a newline everytime it is called (lazy evaluation)
    - note: when read, or readlines are used, the text is completely read and no other reading action can be performed

```python
import matplotlib.pyplot as plt
```

```python
text_file = 'imported data/babaafzal_norm.txt'
with open(text_file, mode='r',encoding='utf-8') as text:
    
    #print(type(text.read()))  # returns the whole text file as a string
    #print(type(text.readlines()))  # retursn the whole text file as a list of strings
    print(type(text.readline()))  # retunrs the next line of the text
    for _ in range(5):
        line_string = text.readline() 
        if len(line_string) > 5: # just a simple quality check
            print(line_string)
```

#### importing pickled data
Pickle is a native pythn data type used to store any variable into serialized byte-type format. It is not human readable.
You can read it fairly similar to text files.

```python
import pickle
# where to save the picke
file_name = 'imported data/picked_dict.pkl'
# first let's create a pickle
family_weights = {'father':145, 'mother':123, 'brother':190}
with open(file_name,mode='wb') as handle:  # write binary
    pickle.dump(family_weights, handle)

# now let's a pickle file
with open(file_name, mode='rb') as file:  # read binary
    data = pickle.load(file)
    
print(data)
```

#### importing excel files
Used a lot in the industry. Usually they are not that big, but we might need some data from them.

```python
#  what is the file name you are looking at
file = r'C:\Users\Amin\Desktop\My job application.xls'
#  read the excel file data (with all the meta data)
data = pd.ExcelFile(file)
# what is the data type
print("Data type is :\t", type(data))
# what are the tabs available in this data set
print("sheet names are:\t" + ", ".join(data.sheet_names) + ".")
# read first tab by index
df1 = data.parse(0)
print(df1.tail(2))
print("*"*50)
# read second tab by index
df2 = data.parse('Jon plan', header=-1)
print(df2.head(2))
```

#### importing SAS and STATA
You need to install and import some packages. 

```python
from sas7bdat import SAS7BDAT

# from here it will be fairly similar to other importing of text files
# except instead of open() function we will be using SAS7BDAT() function

# where the file is located?
file = 'imported data/salary.sas7bdat'
with SAS7BDAT(file) as handle:  # the alias name is not important
    df_sas = handle.to_data_frame()
    
df_sas.head()

df_sas.Y.plot()
_ = plt.xlabel('Age')
_ = plt.ylabel('Income')
```
