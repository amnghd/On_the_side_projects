# Basics of Matplot lib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
housing = pd.read_csv('data/housing.csv')
```

```python
housing.head()
```

### Line plot

Exploring relationship between two variables that have intrisinic monotonics relationship.

```python
room = housing.sort_values('total_bedrooms', ascending = False)
plt.plot(room.total_bedrooms, room.total_rooms)
plt.title('There is a linear relationship between\n\
number of rooms and number of bedrooms')
plt.xlabel('Number of bedrooms in a complex')
plt.ylabel('Number of rooms in a complex')
_ = plt.xticks([0,2000, 4000, 6000], ['0','2k','4k','6k'])
_ =plt.yticks([0,10000, 20000, 30000, 40000], ['0','10k','20k','30k','40k'])
plt.grid(True) # activating grid
plt.text(x = 3500, y= 37000, s= 'linear relationship') # text
```

### Scatter plot
Relationship between two numerical variable. Allows peeking an eye on the correlation between them.
Scatter plot of the location of the houses. Scatter plot of longitude and latitude makes a geospacial graph, that mimics to California (the state that the data is captured from). We can also see that median income is higher in the Bay Area, LA, and San Diego.

```python
sc= plt.scatter(housing.longitude, housing.latitude, \
                alpha = 0.3, c = housing.median_income)
plt.colorbar(sc)
plt.xlabel('Longtitude')
plt.ylabel('Lattitude')
_ = plt.title("Income distribution in the California households")
```

### Histogram
Visualizied EDA on a single numerical variable to get the distribution of variable. Nice overview of how numerical variable is ditributed.
Kernel densitiy estimated is similar to histogram. However, as opposed to histogram, it doesnt use rectangular winows. Rather, it uses gausian windows to calculated the probablity density function of a (random) variable.

From wikipedia:
''Kernel density estimates are closely related to histograms, but can be endowed with properties such as smoothness or continuity by using a suitable kernel. ''

```python
_ = plt.hist(housing.median_income, bins=20, normed=True) # a histogram normed
housing.median_income.plot.kde(bw_method = 0.3) # kde is histogram but windows are not rectangular
plt.title("Histogram and PDF of median income variable.")
plt.xlabel('Income ($k)')
```
